```python
from bs4 import BeautifulSoup
import requests
import pandas as pd
```

```python
url = 'https://www1.bca.gov.sg/bca-directory/registered-contractors/construction?workhead=CW01&grading=All'
```

```python
def scrape(url):
    sheet = []
    page = requests.get(url)
    soup = BeautifulSoup(page.text, 'html')
    wrap = soup.find('div', class_ = 'company-list-wrapper')
    items = wrap.find_all('div', class_ = 'company-list-item')
    for i in items:
        company = i.find('a').text
        uen = i.find('div', class_ = 'uen').text.strip()
        uen = uen[5:]
        address = i.find('div', class_ = 'address').text.strip()
        phone = i.find('div', class_ = 'tel').text.strip()
        fax = i.find('div', class_ = 'fax').text.strip()
        sheet.append([company, uen, address, phone, fax])
    return sheet
```

```python
sheet_0 = scrape(url)
print(sheet_0)
```

```python
sheet_1 = []
for p in range(2,126):
    ext = f'https://www1.bca.gov.sg/bca-directory/registered-contractors/construction?workhead=CW01&grading=All&filter=All&itemsperpage=12&page={p}'
    sheet_1.append(scrape(ext))
print(sheet_1)
```

```python
sheet_list = [inner_list for outer_list in sheet_1 for inner_list in outer_list]
print(sheet_list)
```

```python
df = pd.DataFrame(sheet_0, columns=['Company', 'UEN', 'Address', 'Telephone', 'Fax'])
df.head()
```

```python
df_1 = pd.DataFrame(sheet_list, columns=['Company', 'UEN', 'Address', 'Telephone', 'Fax'])
df_1.head()
```
